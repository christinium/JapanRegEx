### Installation
The installation of this program is desribed very comprehensively on the website: http://ctakes.apache.org/downloads.cgi
- You will need a UMLS® license through NIH.  UMLS stands for Unified Medical Language System and is a standardized way of noting key terminology, classification and coding standards, and associated resources to promote creation of more effective and interoperable biomedical information systems and services, including electronic health records.  There is no charge for licensing any part of the UMLS from NLM. From the website: <br><br>
*There may be separate license fees associated with usage of SNOMED CT which is owned by the International Health Terminology Standards Development Organization (IHTSDO). The SNOMED CT® Affiliate Licence Agreement is included as Appendix 2 of the UMLS Metathesaurus License and outlines possible costs. NLM is a member of IHTSDO and there is no charge for SNOMED CT use in the United States and other member countries.  For information about fees in non-member countries, see the IHTSDO web site: http://www.ihtsdo.org/snomed-ct/get-snomed-ct.*<br><br>
- You will need Java 1.8 or higher (make sure if you have windows, you add it to your path during installation - it should prompt you)
